```python
import matplotlib.pyplot as plt
import numpy as np


def johnson_lindenstrauss_lemma(m, eps):
    """
    Implements the Johnson-Lindenstrauss-Lemma bound.

    :param m: number of data points
    :type m: float
    :param eps: distance resolution parameter
    :type eps: float
    :return: estimate of the lowest possible dimension of that data set
    """

    return 8 * np.log(m) / eps ** 2
```

```python
eps = 0.1
m = np.arange(10, 1010, 10)
```

```python
JL_bound = johnson_lindenstrauss_lemma(m, eps)
```

```python
plt.figure()
for eps in np.linspace(0.1, 0.5, 10):
    plt.plot(m, johnson_lindenstrauss_lemma(m, eps), color="tab:blue")
plt.xlabel("number of data points")
plt.ylabel("JL_bound [minimum dimension]")
plt.yscale("log")
plt.tight_layout()
plt.show()
```
